```python
import matplotlib.pyplot as plt

labels = ["東部", "南部", "北部", "中部"]
sizes = [5, 10, 20, 15]
colors = ["red", "green", "blue", "yellow"]
explode = (0, 0, 0.05, 0)
plt.pie(sizes,explode = explode,labels = labels,colors = colors,\
    labeldistance = 1.1,autopct = "%2.1f%%",shadow = True,\
    startangle = 90,pctdistance = 0.6)
plt.axis("equal")
plt.legend()
plt.show()
```

```python
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6
from bokeh.plotting import figure
from bokeh.transform import factor_cmap

output_file("bar_colormapped.html")

fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
counts = [1,5,7,9,13,16]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p1 = figure(x_range=fruits, plot_height=350, toolbar_location=None, title="Fruit Counts")
p1.vbar(x='fruits', top='counts', width=0.9, source=source, legend="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

p1.xgrid.grid_line_color = None
p1.y_range.start = 0
p1.y_range.end = 20
p1.legend.orientation = "horizontal"
p1.legend.location = "top_center"

show(p1)
```

```python
t=np.arange(0.0,2.0,0.1)
s=np.sin(t*np.pi)
plt.subplot(3,2,1)
plt.plot(t,s,'-',color='black')
plt.ylabel('y1')
plt.subplot(3,2,2)
plt.plot(2*t,s,'^',color='blue')
plt.ylabel('y2')
plt.subplot(3,2,3)
plt.plot(3*t,s,'--',color='red')
plt.ylabel('y3')
plt.subplot(3,2,4)
plt.plot(4*t,s,'*',color='green')
plt.ylabel('y4')

plt.subplot(3,2,5)
plt.plot(5*t,s,'o',color='m')
plt.ylabel('y5')
plt.subplot(3,2,6)
plt.plot(6*t,s,'+',color='k')
plt.ylabel('y6')

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=0.5)
plt.show()
```

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
def f(x,y):
    return np.sin(x**10+np.cos(10+y*x)*np.cos(x))
x=np.linspace(0,5,50)
y=np.linspace(0,5,40)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
#plt.contour(X,Y,Z,colors='black')
Z=f(X,Y)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50,cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
